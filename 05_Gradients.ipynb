# Gradients


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('/content/sample_data/sudoku.jpg',0)
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
display_img(img)
```

```python
#X-axis gradient
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
#Y-axis gradient
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
#Laplacian filter
laplacian = cv2.Laplacian(img,cv2.CV_64F)
```

```python
display_img(sobelx)
```

```python
display_img(sobely)
```

```python
display_img(laplacian)
```
